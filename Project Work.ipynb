```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Data/Cybersecurity Incidents and Data Breaches 2024.csv')
```

```python
industry_counts = df['Industry'].value_counts()
industry_counts.plot(kind='bar', figsize=(10,6), title='Breaches by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Breaches')
plt.show()
```

```python
attack_counts = df["Attack_Type"].value_counts()
attack_counts.plot(kind='bar', figsize=(8,5), title="Breaches by Attack Type")
plt.xlabel("Attack Type")
plt.ylabel("Number of Breaches")
plt.show()
```

```python
breaches_per_day= pd.to_datetime(df['Date_Reported'])
breaches_per_day.plot(kind='line', figsize=(12,6), marker='o', title='Breach Frequency Over Time')
plt.xlabel('Number of Breaches')
plt.ylabel('Date Reported')
plt.show()
```

```python
df.columns = df.columns.str.strip()

company_counts = df.groupby('Company_Name')[['Ransom_Amount_USD', 'Regulatory_Fine_Million_USD']].sum()
company_counts['Total_Amount_USD'] = company_counts['Ransom_Amount_USD'] + company_counts['Regulatory_Fine_Million_USD']
company_counts =company_counts.sort_values(by='Total_Amount_USD', ascending=False)
colors = ['Red','skyblue','Green']
company_counts.plot(kind='bar', figsize=(10,6), title ='How Money Each Company Lost', color=colors)
plt.xlabel("Company")
plt.ylabel("Dollar Amount")
plt.xticks(rotation=85)
plt.show()
```
